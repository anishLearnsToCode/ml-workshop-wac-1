```python
from sklearn.datasets import make_regression 
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,Y=make_regression(n_samples=100,n_features=1,n_informative=1,noise=10.8)
```

```python
plt.scatter(X,Y)
```

```python
Y=Y.reshape((-1,1))
```

# Closed Form Solution

```python
ones=np.ones((X.shape[0],1))
```

```python
X=np.hstack((ones,X))
```

```python
def theta(X,Y):
    return np.dot(np.linalg.pinv(np.dot(X.T,X)),np.dot(X.T,Y))
```

```python
def predict(X,theta):
    return X.dot(theta)
```

```python
thetas=theta(X,Y)
```

```python
Y_pred=predict(X,thetas)
```

```python
plt.scatter(X[:,1],Y)
plt.plot(X[:,1],Y_pred,color='red')
```
