```python
import time
a = np.random.rand(10 ** 8)
b = np.random.rand(10 ** 8)

tick = time.time()
c = np.dot(a, b)
tock = time.time()
print(tock - tick)
```

```python
a = np.random.rand(10 ** 7)
b = np.random.rand(10 ** 7)
tick = time.time();

for index in range(len(a)):
    c += a[index] * b[index]

tock = time.time()    
print(tock - tick)
```

```python
def sigmoid(z):
    return np.exp(-z)

print(sigmoid(0))
print(sigmoid(np.array([0, 1, 2])))
```
